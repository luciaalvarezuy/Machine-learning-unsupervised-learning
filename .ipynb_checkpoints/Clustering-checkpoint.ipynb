```python
import pandas as pd
from sklearn import preprocessing 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from kneed import KneeLocator
import scipy.cluster.hierarchy as shc
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
```

```python
# Cargar los datos
filename = 'C:/Users/Lucía Alvarez/OneDrive - psico.edu.uy/00 Barcelona activa/DF_facebook.csv'
df = pd.read_csv(filename, sep=",", encoding="utf8") 

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Identificamos el número de missing values en cada columna
df.isnull().sum()
```

```python
### Se eliminan las cuatro filas que tienen valores perdidos
df = df.drop(['Column1', 'Column2', 'Column3', 'Column4'], axis=1)

df
```

```python
df.columns
```

```python
#Guardar en un objeto las columnas que vamos a usar en el análisis
df2 = df.drop(['status_id', 'status_type', 'status_published',], axis=1)
df2
```

```python
#Normalizar los datos
scaler = StandardScaler()
df_escalado = scaler.fit_transform(df2)

df_escalado = pd.DataFrame(df_escalado, 
             columns=['num_reactions',
       'num_comments', 'num_shares', 'num_likes', 'num_loves', 'num_wows',
       'num_hahas', 'num_sads', 'num_angrys'])

df_escalado[:5]
```

```python
df_escalado.describe()
```

```python
corr_df = df_escalado.corr(method='pearson').round(2)
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True)
plt.show()
```

Curva elbow (o curva de codo) para explicar cuál sería el valor óptimo de k

```python
kmeans_kwargs = {"init": "random", 
                 "n_init": 10, 
                 "max_iter": 300, 
                 "random_state": 42, }

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_escalado)
    sse.append(kmeans.inertia_)
    
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Número de clústeres")
plt.ylabel("SSE")
plt.show()
```

Para decidir con cuántos clústeres quedamos, se va a realizar un análisis con el paquete kneed con el fin de identificar el punto de codo mediante programación.

```python
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

Mediante el análisis de kneed se sugiere retener 4 clústeres. Es importante recordar que la selección de la cantidad de clúster se realiza mediante la evidencia teórica y análitica que podemos tener de los datos. Es así, a continuación, se estiman tres modelos posibles: tres clusters, cuatro clusteres y cinco clústeres, para posteriormente comparar los indices de ajustes de los tres modelos.

#### Ejercicio 1.  K-means con 3 clusters

```python
#Estimar el modelo de 3 clústeres
model_3 = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42)

#Ajustar el modelo a los datos del estudio:
model_3.fit(df_escalado)
pred = model_3.predict(df_escalado)

#Ver el recuento de valores en cada uno de los clusters
frame = pd.DataFrame(df_escalado)
frame['cluster_3'] = pred
frame['cluster_3'].value_counts()[[0,1,2]]
```

En este análisis tenemos 372 casos que pertenecen al grupo 1 (índice 0), 273 casos pertenecen al grupo 2 (índice 1) y 6405 que pertenecen al grupo 3 (índice 2)

```python
print("Ubicaciones finales del centroide \n" , 
      model_3.cluster_centers_)
```

#### Ejercicio 1.  K-means con 4 clusters

```python
#Estimar el modelo de 4 clusteres
model_4 = KMeans(
    init="random",
    n_clusters=4,
    n_init=10,
    max_iter=300,
    random_state=42)

#Ajustar el modelo a los datos de este estudio:
model_4.fit(df_escalado)
pred = model_4.predict(df_escalado)

#Ver el recuento de valores en cada uno de los clusters

frame = pd.DataFrame(df_escalado)
frame['cluster_4'] = pred
frame['cluster_4'].value_counts()[[0,1,2,3]]
```

En este análisis tenemos 372 puntos que pertenecen al grupo 1 (índice 0), 34 puntos pertenecen al grupo 2, 372 pertenecen al grupo 3 y 6272 pertenecen al grupo 4.

```python
print("Ubicaciones finales del centroide \n" , 
      model_4.cluster_centers_)
```

#### Ejercicio 1.  K-means con 5 clusters

```python
model_5 = KMeans(
    init="random",
    n_clusters=5,
    n_init=10,
    max_iter=300,
    random_state=42)

model_5.fit(df_escalado)
pred = model_5.predict(df_escalado)

frame = pd.DataFrame(df_escalado)
frame['cluster_5'] = pred
frame['cluster_5'].value_counts()[[0,1,2,3,4]]
```

En este análisis tenemos 17 casos que pertenecen al grupo 1, 372 puntos pertenecen al grupo 2, 359 pertenecen al grupo 3, 6269 pertenecen al grupo 4 y 33 al grupo 5.

```python
print("Ubicaciones finales del centroide \n" , 
      model_5.cluster_centers_)
```

### Ejercicio 2
Clasifica los distintos registros utilizando el algoritmo de clustering jerárquico.

```python
df_escalado.head()
```

```python
df_escalado = df_escalado.drop(['cluster_3','cluster_4', 'cluster_5'], axis=1)
```

A continuación se realiza un dendograma

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_escalado, method='ward'))
```

Para poder visualizar los posibles grupos, se realiza un corte en el dendograma:

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_escalado, method='ward'))
plt.axhline(y=120, color='r', linestyle='--')
```

Tenemos tres grupos ya que esta línea corta el dendrograma en tres puntos. Ahora apliquemos el agrupamiento jerárquico para 3 clústeres:

```python
clustering_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
clustering_model.fit(df_escalado)
clustering_model.labels_
```

```python
data_labels = clustering_model.labels_
plt.figure(figsize=(14,10))
sns.scatterplot(x='num_reactions', 
                y='num_comments', 
                data=df_escalado, 
                hue=data_labels).set_title('Gráfico')
```

En el gráfico anterior se observan los datos agrupados en 3 clústeres, cada grupo está identificado con un color diferente.
Al observar este gráfico se observa que:
- El grupo con etiqueta 0 (con color rosado claro) son los que tienen mayor dispersión en relación con el resto de los clústeres. Este grupo se caracteriza por tener elevado número de comentarios y elevado número de reacciones en Facebook

- El grupo con etiqueta 1 (con color violeta claro) son los que reportan baja cantidad de comentarios y moderada cantidad de reacciones

- Por último, el grupo con etiqueta 3 (con color violeta oscuro) son los que reportan bajo número de comentarios y bajo número de reacciones.

En suma, se observa que los 3 clústeres obtenidos reportan diferencias en su comportamiento.


### Ejercicio 3
Calcula el rendimiento del clustering mediante un parámetro como puede ser silueta.

```python

silhouette_coefficients = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df_escalado)
    score = silhouette_score(df_escalado, kmeans.labels_)
    silhouette_coefficients.append(score)
    
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Número de clústeres")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
# Criterio de Inercia
print("Inercia 3 clústeres: ",model_3.inertia_)
print("Inercia 4 clústeres: ",model_4.inertia_)
print("Inercia 5 clústeres: ",model_5.inertia_)
```

En suma, en este estudio se observa:

    - Según el coeficiente de silueta, el modelo con mejor ajuste es el de 5 clústeres
    
    - Teniendo en cuenta el clúster jerárquico, el mejor modelo es el de 3 clústeres
